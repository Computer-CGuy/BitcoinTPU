```python
import time
from pycryptosat import Solver
```

```python

solver = Solver()

with open("./1-Nonce-SAT.cnf", "r") as x:
    for i,line in enumerate(x):
        line = line.strip()
        # if(line[0]=="p" or line[0]=="c"):
        if "p" in line or "c" in line:
            continue

        out = [int(x) for x in line.split()[:-1]]
        # solver.add_clause(out)
#         print(out)
        if(i==655243):
            main_clause=out
        else:
            solver.add_clause(out)
```

```python
for i in range(len(main_clause)-1):
    main_clause[i] = -main_clause[i]
```

```python
### 1st Attemo
# SAT
# Genesis Block
# %%timeit
# SATIABLE = [-2887,2888,-2889,-2890,-2891,2892,2893,2894,2895,2896,2897,2898,2899,2900,2901,2902,-2903,-2904,-2905,-2906,-2907,-2908,-2909,-2910,-2911,-2912,-2913,-2914,-2915,-2916,-2917,-2918,135795]
start = time.time()
res, isSAT = solver.solve(main_clause)
# res, isSAT = solver.solve()
done = time.time()
elapsed = done - start
print(elapsed)
print(res,isSAT)
```
